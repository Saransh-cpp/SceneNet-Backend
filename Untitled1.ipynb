```python
import tensorflow as tf
```

```python
images_ds = tf.data.Dataset.list_files('indoorCVPR_09/*/*/*', shuffle=False)
```

```python
image_count = len(images_ds)
image_count
```

```python
for file in images_ds.take(3):
    print(file.numpy())
```

```python
train_size = int(image_count*0.8)
train_ds = images_ds.take(train_size)
test_ds = images_ds.skip(train_size)
```

```python
def get_label(file_path):
    import os
    parts = tf.strings.split(file_path, os.path.sep)
    return parts[-2]
```

```python
get_label("indoorCVPR_09\\Images\\airport_inside\\airport_inside_0001.jpg")
```

```python
def process_image(file_path):
    label = get_label(file_path)
    img = tf.io.read_file(file_path) # load the raw data from the file as a string
    img = tf.image.decode_jpeg(img)
    img = tf.image.resize(img, [224, 224])
    return img, label
```

```python
img, label = process_image("indoorCVPR_09\\Images\\airport_inside\\airport_inside_0001.jpg")
img.numpy()[:2], label
```

```python
train_ds = train_ds.map(process_image)
test_ds = test_ds.map(process_image)
```

```python
def scale(image, label):
    return image/255, label
```

```python
train_ds = train_ds.map(scale)
```

```python
import cv2
from glob import glob
from PIL import Image
from keras.models import Sequential
from keras.applications.vgg19 import VGG19
from keras.layers import Input, Lambda, Dense, Flatten
from keras.preprocessing.image import ImageDataGenerator
```

```python
vgg19 = VGG19(input_shape=(224, 224, 3), weights='imagenet', include_top=False)
vgg19.trainable = False

flatten_layer = Flatten()
prediction_layer = Dense(67, activation='softmax')

model = Sequential([
    vgg19,
    flatten_layer,
    prediction_layer
])
```

```python
model.summary()

# tell the model what cost and optimization method to use
model.compile(
    loss='categorical_crossentropy',
    optimizer='adam',
    metrics=['accuracy'],
)
```

```python
model.fit(
    train_ds,
    validation_data=test_ds,
    epochs=5,
    batch_size=32,
)
```
